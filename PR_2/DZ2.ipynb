Построим фигуру, состоящую из четырёх графиков. Поскольку мы хотим, чтобы у нашей фигуры было 2 "строки" и 2 "столбца", массив `ax` будет двумерным. Применив к нему метод `.flatten`, мы получим новый массив, в котором сначала будут идти все графики `Axes` из первой строки, а затем все графики из второй строки:
